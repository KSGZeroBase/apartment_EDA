```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(11,11)})

family = "Malgun Gothic" # win
# family = "AppleGothic" # Mac
plt.rc('font', family=family)
```

```python
park_df = pd.read_csv('data/서울특별시/park_df.csv', index_col=0)
apartment_df_by_dong = pd.read_csv('data/서울특별시/apartment_df_by_dong.csv')
```

```python
park_df.dropna()
```

```python
park_df.rename(columns={"park_cultural_facitiy" : "park_cultural_facility"}, inplace=True)
```

```python
apartment_df_by_dong
```

```python
park_df
```

```python
park_df_with_price = pd.merge(park_df, apartment_df_by_dong, on='dong')
park_df_with_price
```

```python
corr_df = park_df_with_price.corr()
corr_df
```

```python
sns.heatmap(data=corr_df, annot=True, cmap="RdYlBu_r", vmin=-1, vmax=1)
plt.show()
```

```python
df_grouped = park_df.groupby('dong').agg(
    oldest_open_year = pd.NamedAgg(column='park_open_year', aggfunc='min'),
    recent_open_year = pd.NamedAgg(column='park_open_year', aggfunc='max'),
    average_park_area = pd.NamedAgg(column='park_area', aggfunc='mean'),
    total_park_area = pd.NamedAgg(column='park_area', aggfunc='sum'),
    park_count = pd.NamedAgg(column='park_area', aggfunc='count'),
)

df_grouped.head(30)
```

```python
df_grouped = df_grouped.reset_index()
```

```python
corr_df = pd.merge(apartment_df_by_dong, df_grouped).corr()
corr_df
```

```python
sns.heatmap(data=corr_df, annot=True, cmap="RdYlBu_r", vmin=-1, vmax=1)
plt.show()
```

```python
np.unique(park_df['park_type'])
```

```python
park_type_pivot = park_df.pivot_table('city', 'dong', 'park_type', aggfunc='count').fillna(0)
park_type_pivot
```

```python
df_grouped.set_index('dong', inplace=True)
```

```python
park_dong_groupby = pd.concat([park_type_pivot, df_grouped], axis=1)
park_dong_groupby
```

```python
apartment_df_by_dong.set_index('dong', inplace=True)
```

```python
concat_df = pd.concat([apartment_df_by_dong, park_dong_groupby], axis=1)
concat_df
```

```python
corr_df = concat_df.corr()
corr_df
```

```python
sns.heatmap(data=corr_df, annot=True, fmt=".2f" ,cmap='RdYlBu_r', vmin=-1, vmax=1)
plt.show()
```
